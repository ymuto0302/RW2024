## 架空のデータセットを生成するコード
乱数を用いているため，実行するたびに「生成されるデータ」は異なる。

```python
import pandas as pd
import numpy as np

# データセットの作成
np.random.seed(0)
data = {
    'A': np.random.normal(50, 10, 100),
    'B': np.random.normal(30, 5, 100),
    'C': np.random.choice(['X', 'Y', 'Z'], 100),
    'D': np.random.normal(100, 20, 100)
}

# 欠損値の挿入
data['A'][np.random.choice(100, 10, replace=False)] = np.nan
data['B'][np.random.choice(100, 5, replace=False)] = np.nan

# 異常値の挿入
data['D'][np.random.choice(100, 5, replace=False)] = data['D'][np.random.choice(100, 5, replace=False)] * 10

# データの DataFrame 化
df = pd.DataFrame(data)
```
